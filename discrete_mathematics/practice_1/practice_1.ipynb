```python
file_text = """Жил-был в одном далёком королевстве добрый и справедливый король. Его народ любил его за мудрость и заботу. Король всегда стремился к тому, чтобы в королевстве царили мир и благополучие.
Однажды в королевство пришёл странный чужеземец. Он привёз с собой удивительные рассказы о других странах и далёких землях. В этих рассказах говорилось о больших городах, могучих замках и необычных существах, которых никто никогда раньше не видел.
Король был очарован этими историями и пригласил чужеземца к себе в замок, чтобы услышать больше. Народ же начал волноваться — ведь этот человек был не таким, как все, и его рассказы были слишком необычными. Некоторые люди считали, что чужеземец принесёт в королевство беду.
Прошло время, и король принял важное решение. Он решил отправиться в путешествие, чтобы самому увидеть те страны и чудеса, о которых рассказывал чужеземец. Народ простился с королём, желая ему удачи и благополучного возвращения."""
```

```python
import re

text = re.sub("[^a-zа-я]", "", file_text.lower())
print(text)
```

```python
import json

single_frequency: dict[str, float] = {}
for symbol in text:
    if single_frequency.get(symbol) is None:
        single_frequency[symbol] = 0
    single_frequency[symbol] += 1

for key in single_frequency.keys():
    single_frequency[key] = single_frequency[key] / len(text)
    single_frequency = dict(sorted(single_frequency.items(), key=lambda item: item[1]))
print(
    f"Однобуквенная частота: {json.dumps(single_frequency, indent=4, ensure_ascii=False)}"
)
```

```python
double_frequency: dict[str, float] = {}
for i in range(len(text) - 1):
    symbols = text[i : i + 2]
    if double_frequency.get(symbols) is None:
        double_frequency[symbols] = 0
    double_frequency[symbols] += 1

for key in double_frequency.keys():
    double_frequency[key] = double_frequency[key] / len(text)
    double_frequency = dict(sorted(double_frequency.items(), key=lambda item: item[1]))
print(
    f"Двухбуквенная частота: {json.dumps(double_frequency, indent=4, ensure_ascii=False)}"
)
```

```python
import math

single_entropy = 0
for value in single_frequency.values():
    single_entropy -= value * math.log2(value)
print(f"Однобуквенная энтропия: {single_entropy}")
```

```python
double_entropy = 0
for value in double_frequency.values():
    double_entropy -= value * math.log2(value)
print(f"Двухбуквенная энтропия: {double_entropy}")
```

```python
code_length = math.log2(len(text))
redundancy = code_length - single_entropy
count_to_remove = int(len(single_frequency) * 0.2)
```

```python
single_frequency_first = single_frequency.copy()
first_keys_to_remove = list(single_frequency_first.keys())[:count_to_remove:]

for key in first_keys_to_remove:
    del single_frequency_first[key]

print(
    f"Однобуквенная частота без 20% наиболее часто встречающихся символов: {json.dumps(single_frequency_first, indent=4, ensure_ascii=False)}"
)
```

```python
first_single_entropy = 0
for value in single_frequency_first.values():
    first_single_entropy -= value * math.log2(value)
print(
    f"Однобуквенная энтропия без 20% наиболее часто встречающихся символов: {first_single_entropy}"
)
```

```python
single_frequency_second = single_frequency.copy()
second_keys_to_remove = list(single_frequency_second.keys())[-count_to_remove:-1:]

for key in second_keys_to_remove:
    del single_frequency_second[key]

print(
    f"Однобуквенная частота без 20% наиболее редко встречающихся символов: {json.dumps(single_frequency_second, indent=4, ensure_ascii=False)}"
)
```

```python
second_single_entropy = 0
for value in single_frequency_second.values():
    second_single_entropy -= value * math.log2(value)
print(
    f"Однобуквенная энтропия без 20% наиболее редко встречающихся символов: {second_single_entropy}"
)
```
