```python
# 4.3.1
import numpy as np
import matplotlib.pyplot as plt

array = np.random.rand(100)
mean_value = np.mean(array)
median_value = np.median(array)
print(f"Среднее значение: {mean_value}")
print(f"Медианное значение: {median_value}")
plt.scatter(range(len(array)), array)
plt.title('Точечная диаграмма рассеяния')
plt.xlabel('Индекс')
plt.ylabel('Значение')
plt.show()
```

```python
# 4.3.2

import numpy as np
import matplotlib.pyplot as plt

def my_function(x):
    return (np.sqrt(1 + np.exp(np.sqrt(x)) + np.cos(x)**2)) / np.abs(1 - np.sin(x)**3) + np.log(2*x)

x_values = np.arange(1, 11)

y_values = my_function(x_values)

half_slice = y_values[:len(y_values)//2]

plt.figure(figsize=(10, 6))

plt.subplot(3, 1, 1)
plt.plot(x_values, y_values, label="Полный массив", color='blue')
plt.title("Линейный график для всего массива")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()

plt.subplot(3, 1, 2)
plt.scatter(x_values[:len(half_slice)], half_slice, color='red', label="Срез первой половины")
plt.title("Точечный график для первой половины массива")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()

plt.subplot(3, 1, 3)
plt.plot(x_values[:len(half_slice)], half_slice, color='green', label="Линейный срез первой половины")
plt.title("Линейный график для среза первой половины")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# 4.3.3

import numpy as np
import matplotlib.pyplot as plt
from numpy import trapz

def my_function(x):
    return np.abs(np.cos(x * np.exp(np.cos(x) + np.log(x + 1))))

x_values = np.arange(0.1, 10, 1)
y_values = my_function(x_values)

plt.plot(x_values, y_values, label="График функции")
plt.fill_between(x_values, y_values, color="skyblue", alpha=0.4)
plt.title("График функции с заливкой площади")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()

plt.show()

area_trapz = trapz(y_values, x_values)
print(f"Площадь под графиком (по правилу трапеций): {area_trapz}")

```

```python
# 4.3.4

import yfinance as yf
import matplotlib.pyplot as plt

apple = yf.download('AAPL', start='2021-01-01', end='2021-12-31')
microsoft = yf.download('MSFT', start='2021-01-01', end='2021-12-31')
google = yf.download('GOOGL', start='2021-01-01', end='2021-12-31')

plt.figure(figsize=(10, 6))

plt.subplot(3, 1, 1)
plt.plot(apple.index, apple['Close'], label='Apple', color='blue')
plt.title('Стоимость акций Apple в 2021 году')
plt.ylabel('Цена (USD)')
plt.grid(True)

plt.subplot(3, 1, 2)
plt.plot(microsoft.index, microsoft['Close'], label='Microsoft', color='green')
plt.title('Стоимость акций Microsoft в 2021 году')
plt.ylabel('Цена (USD)')
plt.grid(True)

plt.subplot(3, 1, 3)
plt.plot(google.index, google['Close'], label='Google', color='red')
plt.title('Стоимость акций Google в 2021 году')
plt.ylabel('Цена (USD)')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# 4.3.5

import math

def calculator(x, y):
    print(f"Сложение: {x} + {y} = {x + y}")
    print(f"Вычитание: {x} - {y} = {x - y}")
    print(f"Умножение: {x} * {y} = {x * y}")
    if y != 0:
        print(f"Деление: {x} / {y} = {x / y}")
    else:
        print("Деление на ноль невозможно")
    
    print(f"e^(x + y): {math.exp(x + y)}")
    print(f"sin(x + y): {math.sin(x + y)}")
    print(f"cos(x + y): {math.cos(x + y)}")
    print(f"x^y: {x ** y}")

x = float(input("Введите значение x: "))
y = float(input("Введите значение y: "))
calculator(x, y)
```
